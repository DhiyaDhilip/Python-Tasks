```python
# ✅ 1. Factorial of a given number

n=int(input("Enter a number"))
fact=1
for i in range(1,n+1):
    fact*=i
print("Factorial=",fact)
```

```python
# ✅ 2. Given number is prime or not

n=int(input("Enter a number"))
if n<=1:
    print(f"{n} is not a prime number")
else:
    for i in range(2,n):
        if n%i==0:
            print(f"{n} is not a prime number")
            break
    else:
        print(f"{n} is a Prime number")
```

```python
# ✅ 3. Fibonacci series

n=int(input("Enter a number"))
a,b=0,1
count=0
while count<n:
    print(a,end=" ")
    a,b=b,a+b
    count+=1

for i in range(n):
    print(a, end=" ")
    a, b = b, a+b
```

```python
# ✅ 4. Palindrome

s = input("Enter string/number: ")
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# ✅ 5. Sum of digits

n=input("Enter a number")
sum=0
for i in n:
    sum+=int(i)
print("Sum of digits is",sum)
```

```python
# ✅ 6. Print alternative values in a given string

s = input("Enter string: ")
print(s)
print(s[::2])
```

```python
# ✅ 7. Concatenate 2 string lists

l1=input().split()
l2=input().split()
add=[]
for i in range(len(l1)):
    add.append(l1[i]+l2[i])
print(add)

list1 = input().split()
list2 = input().split()

result = list(map(lambda x, y: x + y, list1, list2))
print(result)
```

```python
# ✅ 8. Access list elements using indices list

elements=input().split()
print(elements)
indices=list(map(int, input().split()))
print(indices)

give=[]
for i in indices:
    give.append(elements[i])
print(give)
```

```python
# ✅ 9. Largest & smallest in list

lst = list(map(int, input().split()))
print(lst)
print("Smallest:", min(lst))
print("Largest:", max(lst))

sort=sorted(lst)
print("Smallest:", sort[0])
print("Largest:", sort[-1])
```

```python
# ✅ 10. Sort list (ascending & descending)
lst=list(map(int, input().split()))
print(lst)
asc=lst[:]
asc.sort()
dsc = lst[:]
dsc.sort(reverse=True)
print("Ascending:", asc)
print("Descending:", dsc)

lst = list(map(int, input().split()))
print("Ascending:", sorted(lst))
print("Descending:", sorted(lst, reverse=True))
```

```python
# ✅ 11. Count vowels in each word
words=input().split()
vowels="aeiouAEIOU"

result=[]
count_of_vowels=[]
for w in words:
    count=0
    for c in w:
        if c in vowels:
            count+=1
            result.append(c)
    count_of_vowels.append(count)
print(result)
print(count_of_vowels)
```

```python
# ✅ 12. Count occurrences of an element in list

lst = input("Enter list elements: ").split()
print(lst)
item = input("Enter element to count: ")
print(item)
print("Count =", lst.count(item))

count = 0
for i in lst:
    if i == item:
        count += 1
print("Count =", count)
```

```python
# ✅ 13. Find N largest elements

lst = list(map(int, input("Enter numbers: ").split()))
print(lst)
n = int(input("Enter N: "))
print(n)

largest = sorted(lst, reverse=True)[:n]
print("N largest elements:", largest)
# or
lst.sort(reverse=True)
print(lst[:n])
```

```python
# ✅ 14. Insert element in sorted list
lst = list(map(int, input().split()))
print(lst)
x = int(input("Enter element to insert: "))
print("Inserted element:", x)

lst.append(x)
lst.sort()
print(lst)
```

```python
# ✅ 15. Cubes of n integers (list comprehension)

nums = list(map(int, input("Enter numbers separated by space: ").split()))
cubes = [x**3 for x in nums]
print("Cubes:", cubes)
```

```python
# ✅ 20. Mask credit card except last 4
def hide_card(card):
    return "*" * (len(card) - 4) + card[-4:]

card = input("Enter card number: ")
print(hide_card(card))
```
